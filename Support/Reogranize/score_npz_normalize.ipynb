```python
import numpy as np

npy_path = "C:\\Users\\jiayang\\ipynb\\APS_Thesis\\data\\imagenet_count.npy"
npy_file = np.load(npy_path)

# cumulate real probability
row_sums = npy_file.sum(axis=1, keepdims=True)

# avoid dividing 0 (0/0 -> 0/1)
row_sums[row_sums == 0 ] = 1

# normalize
# e.g. {0, 0, 0.9, ..., 0.9, 0} -> {0, 0, 0.5,..., 0.5, 0}
imagenet_count_normalize = npy_file / row_sums

print("Matrix shape:", imagenet_count_normalize.shape)
row_index = 99
print(f"Row {row_index} content:", imagenet_count_normalize[row_index])
```

```python
save_path = "C:\\Users\\jiayang\\ipynb\\APS_Thesis\\data\\imagenet_count_normalize.npy"
np.save(save_path, imagenet_count_normalize)
print(f"npy file saved at {save_path}")
```
